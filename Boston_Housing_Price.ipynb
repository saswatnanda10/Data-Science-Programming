```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.DataFrame(pd.read_csv('Boston_Test.csv'))
df1=pd.DataFrame(pd.read_csv('Boston_Train.csv'))
dataset=pd.merge(df,df1,how="outer")
#common=pd.merge(df,df1,how="inner")
dataset.drop('Unnamed: 0', axis=1, inplace=True)
dataset

# Reading and merging datasets
```

```python
y=dataset['medv']
X=dataset.drop('medv',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=4)
```

**Linear Regression** 

```python
from sklearn.linear_model import LinearRegression

l=LinearRegression()
l.fit(X_train,y_train)
#Training the model
```

```python
l.intercept_
#calculating the y-intercept
```

```python
y_pred=l.predict(X_train)
y_pred
```

```python
from sklearn import metrics
print("R^2 score=",metrics.r2_score(y_train,y_pred))
```

```python
plt.scatter(y_train,y_pred)
plt.xlabel('Prices')
plt.ylabel('Predicted Prices')
plt.title('Prices vs Predicted Prices')
```

```python
plt.title('Prices vs Remaining')
plt.scatter(y_pred,y_train-y_pred)
```

```python
from sklearn import metrics
y_test_pred=l.predict(X_test)
acc_linreg = metrics.r2_score(y_test, y_test_pred)
acc_linreg
```
